### The expand and n Parameters of the str.split() method

```python
chicago = pd.read_csv('../datasets/chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.head()
```

```python
chicago[['First Name', 'Last Name']] = chicago['Name'].str.split(',', expand=True, )
chicago.head(3)
```

```python
# If you try to split phrases with different num of spaces you end up with this
# large df
chicago['Position Title'].str.split(' ', expand=True)
```

```python
# Limit number of splits to have only 2 columns
# use n param
# This will put the first word in the 1st col and the
# rest of the phrase into the 2nd col
chicago['Position Title'].str.split(' ', expand=True, n=1)
```

```python
chicago[['Fist Title Word', 'Seconds Title Word']] = chicago['Position Title'].str.split(' ', expand=True, n=1)
chicago.head(3)
```
