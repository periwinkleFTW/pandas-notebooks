### Filter Dataframe with more than one condition

```python
df = pd.read_csv('../datasets/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head()
```

```python
# AND operator

mask1 = df['Gender'] == 'Male'
mask2 = df['Team'] == 'Marketing'

# In Pandas 'and' is '&'
df[mask1 & mask2]
```

```python
# OR operator

mask1 = df['Senior Management'] 
mask2 = df['Start Date'] < "1990-01-01"

# In Pandas 'or' is '|'
df[mask1 | mask2]
```

### .isin Method

```python
# Checking for multiple values
# Extract rows with teams Legal, Sales, Product
mask = df['Team'].isin(['Legal', 'Sales', 'Product'])
df[mask]
```

### The .isnull and .notnull Method

```python
mask = df['Team'].isnull()

df[mask]
```

### The .between() method

```python
# Both values are INCLUSIVE!!!
df[df['Salary'].between(60000, 70000)]
```

### Checking for duplicate rows with .duplicate() method

```python
df = pd.read_csv('../datasets/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.sort_values('First Name', inplace=True)
df.head()
```

```python
# It marks first occurrence as unique and all others as dupes, careful
# This returns only names that occur exactly once
df[~df['First Name'].duplicated(keep=False)]
```

### Delete duplicate rows with drop_duplicates() method

```python
len(df)
```

```python
# Removes rows where ALL values are duplicate 
df.drop_duplicates()
```

```python
# You have to be careful with this method. Only returns rows with values
# that occur exactly once. You might get an empty df
df.drop_duplicates(subset=['First Name'], keep=False)
```
