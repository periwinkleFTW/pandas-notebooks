### Broadcasting operations

```python
nba = pd.read_csv('../datasets/nba.csv')
nba
```

```python
# Broadcasting means operations are applied to each value individually
# instead of applying it to the whole series
nba['Age'].add(5)
nba['Age'] + 5

nba['Salary'].sub(100000)
nba['Salary'] - 100000

nba['Weight'].mul(0.453592)
nba['Weight in Kilograms'] = nba['Weight'] * 0.453592
nba
```

### Drop rows with Null values

```python
# Drops all rows with at least one Null value
nba.dropna()
```

```python
# Drops all rows with all Null values
nba.dropna(how='all')
```

```python
# Drops columns with at least one Null value
nba.dropna(axis=1)
```

```python
# Remove row if it has Null in a specified column
nba.dropna(subset=['Salary'])
```

### Fill in Null values with .fillna() method

```python
# This will replace all values with 0
nba.fillna(0)
```

```python
nba['Salary'].fillna(0, inplace=True)
nba['College'].fillna('No College', inplace=True)
nba
```

### Convert datatypes of the Series with .astype()

```python
nba = pd.read_csv('../datasets/nba.csv').dropna(how='any')
nba['Salary'].fillna(0, inplace=True)
nba['College'].fillna('No College', inplace=True)
nba.head()
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
nba['Salary'] = nba['Salary'].astype('int')
nba.head()
```

```python
nba['Number'] = nba['Number'].astype('int')
nba['Age'] = nba['Age'].astype('int')
nba.head()
```

```python
# Pandas datatype - category
# Saves a lot of memory
nba['Position'] = nba['Position'].astype('category')
nba['Team'] = nba['Team'].astype('category')
nba.info()
```

### Sort Dataframe with .sort_values() method

```python
nba = pd.read_csv('../datasets/nba.csv')
nba.head()
```

```python
nba.sort_values('Name', na_position='first')
```

```python
nba.sort_values(['Team','Name'], ascending=[True, False])
```

### Sort Dataframe with .sort_index() method

```python
nba.sort_index(axis=0)
```

### Rank values with .rank() method

```python
nba.dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')
nba.head()
```

```python
nba.sort_values('Salary', ascending=False)
```
