### Create Multi Index with set_index method

```python
# Best practice is to put the col with least num of unique
# values as outer layer

# Indexed by Date and then by Country
bigmac.set_index(keys=['Date', 'Country'], inplace=True)
bigmac.head(3)
```

```python
# Sort will first sort the outer layer, then layers below
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
# This will return a tuple for multi-index dfs
bigmac.index
```

```python
# This will return index cols names
bigmac.index.names
```

### Extrac index level values with get_level_values method

```python
# Create multi index during import
bigmac = pd.read_csv('../datasets/bigmac.csv', 
                     parse_dates=['Date'], 
                     index_col=['Date', 'Country'])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
# Call this method on INDEX!!!
# these calls are equivalent
bigmac.index.get_level_values('Date')
bigmac.index.get_level_values[0]
```

```python
bigmac.index.get_level_values('Country')
bigmac.index.get_level_values[1]
```

### Change index level name with set_names method

```python
# set_names is invoked on index!!!
bigmac.index.set_names(names=['Day', 'Locations'], inplace=True)
bigmac.head()
```

```python
bigmac.index.set_names(names='Date', 
                       level=0, 
                       inplace=True)
# Same result
bigmac.index.set_names(names='Date', 
                       level='Day', 
                       inplace=True)
bigmac.head()
```

### The sort_index method on a multi index df

```python
bigmac = pd.read_csv('../datasets/bigmac.csv', 
                     parse_dates=['Date'], 
                     index_col=['Date', 'Country'])
bigmac.head()
```

```python
# Sort 1st index in ascending, 2nd in descending
# returns a brand new df
bigmac.sort_index(ascending=[True, False], inplace=True)
bigmac.head(3)
```

```python
# Sort only a particular index, use level
bigmac.sort_index(level=0)
bigmac.sort_index(level='Date')
```

```python
bigmac.sort_index(level=1)
bigmac.sort_index(level='Country')
```

### Extrac rows from a multi index df using .loc (for labels) and .iloc (for indices)

```python
bigmac = pd.read_csv('../datasets/bigmac.csv', 
                     parse_dates=['Date'], 
                     index_col=['Date', 'Country'])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
# Have 1st arg encapsulate row info, and 2nd col info
bigmac.loc[('2010-01-01', 'Argentina'), 'Price in US Dollars']
```

```python
# Multi index does not affect .iloc
bigmac.iloc[0]

bigmac.iloc[[10, 20, 30]]
```

### The .transpose method on a multi index df

```python
# Swap rows and columns
# Does not mutate df
# If you want to make it permanent - assign to a new df
bigmac = bigmac.transpose()
bigmac.head()
```

```python
# Extract prices for a specified date from Sri Lanka to Ukraine
bigmac.loc[('Price in US Dollars',), ('2010-01-01', 'Sri Lanka'):('2010-01-01','Ukraine')]
```

### The .swaplevel() method

```python
bigmac = pd.read_csv('../datasets/bigmac.csv', 
                     parse_dates=['Date'], 
                     index_col=['Date', 'Country'])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
# Swap index positions
bigmac.swaplevel('Date', 'Country')
bigmac.swaplevel(0, 1)
```

### The .stack() method

```python
# Takes the column labels and moves them to the left
# Increases num of rows and reduces num of cols
world = pd.read_csv('../datasets/worldstats.csv', index_col=['country', 'year'])
world.head()
```

```python
world.stack()
```

```python
world.stack().to_frame()
```

### The .unstack() method

```python
# Takes the most inner index layer and makes it a column
world = pd.read_csv('../datasets/worldstats.csv', index_col=['country', 'year'])
world.head()
```

```python
s = world.stack()
s.head()
```

```python

s.unstack().unstack()
```

```python
s.unstack().unstack().unstack()
```

```python
s.unstack(-1)
```

```python
# Make year (1st) and country (2nd) cols
s.unstack(level=[1, 0])
# s.unstack(level=['Year', 'Country'])
```

```python
# Fill value replaces NaN
s.unstack('year', fill_value=0)
```

### The .pivot() method

```python
sales = pd.read_csv('../datasets/salesmen.csv', parse_dates=['Date'])
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head()
```

```python
# Condense df, 5 rows become 1
# Salesman values become column header
sales.pivot(index='Date', columns='Salesman', values='Revenue')
```

### The .pivot_table() method

```python
foods = pd.read_csv('../datasets/foods.csv')
foods.head()
```

```python
# Get sum spent based on gender and item per city
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc='sum')
```

### The pd.melt() method to create narrow dataset (opposite of pivot)

```python
sales = pd.read_csv('../datasets/quarters.csv')
sales.head()
```

```python
# Move column headers into rows
# Condense several columns into a single col/row
pd.melt(sales, id_vars='Salesman', var_name='Quarter', value_name='Revenue')
```
